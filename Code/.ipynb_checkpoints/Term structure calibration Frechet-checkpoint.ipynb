```python
# Price_po = np.vectorize(price_po) # price func
Price_f = np.vectorize(price_f) # price func
Impliedvol = np.vectorize(impliedvol_func) # implied vol func
```

```python
def sigma_model_f(x, s, k, t, rd, rf):
    """
    f!
    
    x[0]: sigma0
    x[1]: sigma1
    x[2]: alpha
    x[3]: kappa
    x[4]: scale
    
    """
    # price_po is from fastcalib.py
    ########
    price = Price_f(s, k, t, rd, rf, x[0], x[1], x[2], x[3], x[4])
    ###
    
    # impliedvo_func is from jitbs.py
    # Impliedvol = np.vectorize(impliedvol_func)
    return Impliedvol(s, k, t, rd, rf, price) 
```

```python
# fit with sigma lower bar
def Fit_BF_sigmalow(OBJ, INPUT, GUESS, S, T, Rd, Rf, Verbose = 2):
    
    # low bd
    bd1l = [np.minimum(0.0001, OBJ[2+5*0]*0.3), OBJ[2+5*0], -0.8]
    bd2l = [np.minimum(0.0001, OBJ[2+5*1]*0.3), OBJ[2+5*1], -0.8]
    bd3l = [np.minimum(0.0001, OBJ[2+5*2]*0.3), OBJ[2+5*2], -0.8]
    bd4l = [np.minimum(0.0001, OBJ[2+5*3]*0.3), OBJ[2+5*3], -0.8]
    bd5l = [np.minimum(0.0001, OBJ[2+5*4]*0.3), OBJ[2+5*4], -0.8]
    bd6l = [np.minimum(0.0001, OBJ[2+5*5]*0.3), OBJ[2+5*5], -0.8]

    bd1u = [OBJ[2+5*0],                         2,       0.8]
    bd2u = [OBJ[2+5*1],                         2,       0.8]
    bd3u = [OBJ[2+5*2],                         2,       0.8]
    bd4u = [OBJ[2+5*3],                         2,       0.8]
    bd5u = [OBJ[2+5*4],                         2,       0.8]
    bd6u = [OBJ[2+5*5],                         2,       0.8]
   
        
    BOUNDS = (bd1l+bd2l+bd3l+bd4l+bd5l+bd6l+[0.0001]+[1.0001], 
              bd1u+bd2u+bd3u+bd4u+bd5u+bd6u+[10.001]+[100.0001])
    

    
    # Check that the Guess is correct
    # for sigma upper
    GUESS[1+3*0] = np.maximum(GUESS[1+3*0], OBJ[2+5*0])
    GUESS[1+3*1] = np.maximum(GUESS[1+3*1], OBJ[2+5*1])
    GUESS[1+3*2] = np.maximum(GUESS[1+3*2], OBJ[2+5*2])
    GUESS[1+3*3] = np.maximum(GUESS[1+3*3], OBJ[2+5*3])
    GUESS[1+3*4] = np.maximum(GUESS[1+3*4], OBJ[2+5*4])
    GUESS[1+3*5] = np.maximum(GUESS[1+3*5], OBJ[2+5*5])

    # for sigma lower
    GUESS[0+3*0] = np.maximum(np.minimum(GUESS[0+3*0], OBJ[2+5*0]*0.5), 0.0001)
    GUESS[0+3*1] = np.maximum(np.minimum(GUESS[0+3*1], OBJ[2+5*1]*0.5), 0.0001)
    GUESS[0+3*2] = np.maximum(np.minimum(GUESS[0+3*2], OBJ[2+5*2]*0.5), 0.0001)
    GUESS[0+3*3] = np.maximum(np.minimum(GUESS[0+3*3], OBJ[2+5*3]*0.5), 0.0001)
    GUESS[0+3*4] = np.maximum(np.minimum(GUESS[0+3*4], OBJ[2+5*4]*0.5), 0.0001)
    GUESS[0+3*5] = np.maximum(np.minimum(GUESS[0+3*5], OBJ[2+5*5]*0.5), 0.0001)

    ### check sigma model ###
    Sigma = sigma_model_f

    def fun(x,k,y):
        z = x
        term = np.empty(IND)

        #  k[i]--[KxP, KxP, KATM, KxC, KxC] & y[i]--[SxP, SxP, SATM, SxC, SxC]

        for i in INDLIST[0+5*0 : 5+5*0]: # 3m
#             print(i,POS_ALPHA,POS_SCALE)
            term[i] = (Sigma([z[0+3*0],z[1+3*0],z[POS_ALPHA],z[2+3*0],z[POS_SCALE]], S, k[i],T[0],Rd[0],Rf[0]) - y[i])
        
        for i in INDLIST[0+5*1 : 5+5*1]: # 6m
            term[i] = (Sigma([z[0+3*1],z[1+3*1],z[POS_ALPHA],z[2+3*1],z[POS_SCALE]], S, k[i],T[1],Rd[1],Rf[1]) - y[i])
        
        for i in INDLIST[0+5*2 : 5+5*2]: # 9m
            term[i] = (Sigma([z[0+3*2],z[1+3*2],z[POS_ALPHA],z[2+3*2],z[POS_SCALE]], S, k[i],T[2],Rd[2],Rf[2]) - y[i])
        
        for i in INDLIST[0+5*3 : 5+5*3]: # 12m
            term[i] = (Sigma([z[0+3*3],z[1+3*3],z[POS_ALPHA],z[2+3*3],z[POS_SCALE]], S, k[i],T[3],Rd[3],Rf[3]) - y[i])
        
        for i in INDLIST[0+5*4 : 5+5*4]: # 18m
            term[i] = (Sigma([z[0+3*4],z[1+3*4],z[POS_ALPHA],z[2+3*4],z[POS_SCALE]], S, k[i],T[4],Rd[4],Rf[4]) - y[i])
        
        for i in INDLIST[0+5*5 : 5+5*5]: # 2y
            term[i] = (Sigma([z[0+3*5],z[1+3*5],z[POS_ALPHA],z[2+3*5],z[POS_SCALE]], S, k[i],T[5],Rd[5],Rf[5]) - y[i])
           
        return term

    # run least squares
    res_lsq = least_squares(fun, 
                            GUESS, 
                            loss = 'cauchy', 
                            bounds=BOUNDS, 
                            args = (INPUT, OBJ), 
                            verbose = Verbose,
                            max_nfev=180
                           )
    x = res_lsq.x # 10 elements 
    print('\nQuotes vs Calibration...')

    # param
    x1m = [x[0+3*0], x[1+3*0], x[POS_ALPHA], x[2+3*0], x[POS_SCALE]]
    x2m = [x[0+3*1], x[1+3*1], x[POS_ALPHA], x[2+3*1], x[POS_SCALE]]
    x3m = [x[0+3*2], x[1+3*2], x[POS_ALPHA], x[2+3*2], x[POS_SCALE]]
    x6m = [x[0+3*3], x[1+3*3], x[POS_ALPHA], x[2+3*3], x[POS_SCALE]]
    x9m = [x[0+3*4], x[1+3*4], x[POS_ALPHA], x[2+3*4], x[POS_SCALE]]
    x1y = [x[0+3*5], x[1+3*5], x[POS_ALPHA], x[2+3*5], x[POS_SCALE]]

    # k
    k1m = INPUT[0+5*0 : 5+5*0]
    k2m = INPUT[0+5*1 : 5+5*1]
    k3m = INPUT[0+5*2 : 5+5*2]
    k6m = INPUT[0+5*3 : 5+5*3]
    k9m = INPUT[0+5*4 : 5+5*4]
    k1y = INPUT[0+5*5 : 5+5*5]

    # t,rd,rf
    t1m,rd1m,rf1m = T[0], Rd[0], Rf[0]
    t2m,rd2m,rf2m = T[1], Rd[1], Rf[1]
    t3m,rd3m,rf3m = T[2], Rd[2], Rf[2]
    t6m,rd6m,rf6m = T[3], Rd[3], Rf[3]
    t9m,rd9m,rf9m = T[4], Rd[4], Rf[4]
    t1y,rd1y,rf1y = T[5], Rd[5], Rf[5]

    # s - calib vols
    s1m = Sigma(x1m, S, k1m, t1m, rd1m, rf1m)
    s2m = Sigma(x2m, S, k2m, t2m, rd2m, rf2m)
    s3m = Sigma(x3m, S, k3m, t3m, rd3m, rf3m)
    s6m = Sigma(x6m, S, k6m, t6m, rd6m, rf6m)
    s9m = Sigma(x9m, S, k9m, t9m, rd9m, rf9m)
    s1y = Sigma(x1y, S, k1y, t1y, rd1y, rf1y)

    # quotes
    obj1m = OBJ[0+5*0 : 5+5*0]
    obj2m = OBJ[0+5*1 : 5+5*1]
    obj3m = OBJ[0+5*2 : 5+5*2]
    obj6m = OBJ[0+5*3 : 5+5*3]
    obj9m = OBJ[0+5*4 : 5+5*4]
    obj1y = OBJ[0+5*5 : 5+5*5]

    # show quotes
    print('1m---','\t',np.round(s1m,5),'\t',np.round(obj1m,5))
    print('2m---','\t',np.round(s2m,5),'\t',np.round(obj2m,5))
    print('3m---','\t',np.round(s3m,5),'\t',np.round(obj3m,5))
    print('6m---','\t',np.round(s6m,5),'\t',np.round(obj6m,5))
    print('9m---','\t',np.round(s9m,5),'\t',np.round(obj9m,5))
    print('1y---','\t',np.round(s1y,5),'\t',np.round(obj1y,5))
    
    # all
    s_all = np.empty(IND)
    s_all[0+5*0 : 5+5*0] = s1m 
    s_all[0+5*1 : 5+5*1] = s2m 
    s_all[0+5*2 : 5+5*2] = s3m 
    s_all[0+5*3 : 5+5*3] = s6m 
    s_all[0+5*4 : 5+5*4] = s9m 
    s_all[0+5*5 : 5+5*5] = s1y 

    # abs mean err
    e1m = (np.abs(obj1m - s1m) / obj1m).sum() / 5
    e2m = (np.abs(obj2m - s2m) / obj2m).sum() / 5
    e3m = (np.abs(obj3m - s3m) / obj3m).sum() / 5
    e6m = (np.abs(obj6m - s6m) / obj6m).sum() / 5
    e9m = (np.abs(obj9m - s9m) / obj9m).sum() / 5
    e1y = (np.abs(obj1y - s1y) / obj1y).sum() / 5
    
    # total error
    o_all = OBJ
    e_all=(np.abs(o_all-s_all)/o_all).sum() / IND # Important!45!!!


    return (
        res_lsq.x,
        s1m,s2m,s3m,s6m,s9m,s1y,
        e1m,e2m,e3m,e6m,e9m,e1y,
        e_all
        )
```

```python
def func(row):
    global GUESS
    
    S = row['S']

    # rd, rf, t
    Rd3m,Rf3m,T3m = row['Rd3m'],row['Rf3m'],row['T3m']
    Rd6m,Rf6m,T6m = row['Rd6m'],row['Rf6m'],row['T6m']
    Rd9m,Rf9m,T9m = row['Rd9m'],row['Rf9m'],row['T9m']
    Rd1y,Rf1y,T1y = row['Rd1y'],row['Rf1y'],row['T1y']
    Rd18m,Rf18m,T18m = row['Rd18m'],row['Rf18m'],row['T18m']
    Rd2y,Rf2y,T2y = row['Rd2y'],row['Rf2y'],row['T2y']

    # quotes
    S10P3m,S25P3m,SATM3m,S25C3m,S10C3m = row['S10P3m'],row['S25P3m'],row['SATM3m'],row['S25C3m'],row['S10C3m']
    S10P6m,S25P6m,SATM6m,S25C6m,S10C6m = row['S10P6m'],row['S25P6m'],row['SATM6m'],row['S25C6m'],row['S10C6m']
    S10P9m,S25P9m,SATM9m,S25C9m,S10C9m = row['S10P9m'],row['S25P9m'],row['SATM9m'],row['S25C9m'],row['S10C9m']
    S10P1y,S25P1y,SATM1y,S25C1y,S10C1y = row['S10P1y'],row['S25P1y'],row['SATM1y'],row['S25C1y'],row['S10C1y']
    S10P18m,S25P18m,SATM18m,S25C18m,S10C18m = row['S10P18m'],row['S25P18m'],row['SATM18m'],row['S25C18m'],row['S10C18m']
    S10P2y,S25P2y,SATM2y,S25C2y,S10C2y = row['S10P2y'],row['S25P2y'],row['SATM2y'],row['S25C2y'],row['S10C2y']

    # k
    K10P3m,K25P3m,KATM3m,K25C3m,K10C3m = row['K10P3m'],row['K25P3m'],row['KATM3m'],row['K25C3m'],row['K10C3m']
    K10P6m,K25P6m,KATM6m,K25C6m,K10C6m = row['K10P6m'],row['K25P6m'],row['KATM6m'],row['K25C6m'],row['K10C6m']
    K10P9m,K25P9m,KATM9m,K25C9m,K10C9m = row['K10P9m'],row['K25P9m'],row['KATM9m'],row['K25C9m'],row['K10C9m']
    K10P1y,K25P1y,KATM1y,K25C1y,K10C1y = row['K10P1y'],row['K25P1y'],row['KATM1y'],row['K25C1y'],row['K10C1y']
    K10P18m,K25P18m,KATM18m,K25C18m,K10C18m = row['K10P18m'],row['K25P18m'],row['KATM18m'],row['K25C18m'],row['K10C18m']
    K10P2y,K25P2y,KATM2y,K25C2y,K10C2y = row['K10P2y'],row['K25P2y'],row['KATM2y'],row['K25C2y'],row['K10C2y']
    
    OBJ = np.array([
        S10P3m,S25P3m,SATM3m,S25C3m,S10C3m,
        S10P6m,S25P6m,SATM6m,S25C6m,S10C6m,
        S10P9m,S25P9m,SATM9m,S25C9m,S10C9m,
        S10P1y,S25P1y,SATM1y,S25C1y,S10C1y,
        S10P18m,S25P18m,SATM18m,S25C18m,S10C18m,
        S10P2y,S25P2y,SATM2y,S25C2y,S10C2y,
    ])

    INPUT = np.array([
        K10P3m,K25P3m,KATM3m,K25C3m,K10C3m,
        K10P6m,K25P6m,KATM6m,K25C6m,K10C6m,
        K10P9m,K25P9m,KATM9m,K25C9m,K10C9m,
        K10P1y,K25P1y,KATM1y,K25C1y,K10C1y,
        K10P18m,K25P18m,KATM18m,K25C18m,K10C18m,
        K10P2y,K25P2y,KATM2y,K25C2y,K10C2y,
        ])

    T_ls = [T3m,T6m,T9m,T1y,T18m,T2y]
    Rd_ls = [Rd3m,Rd6m,Rd9m,Rd1y,Rd18m,Rd2y]
    Rf_ls = [Rf3m,Rf6m,Rf9m,Rf1y,Rf18m,Rf2y]


    print(f'\n------------- ------------------------- -------------------- ------------------- -------')
    print(f'Dates now... {row.name}...\n')
    res_lsq_x,s3m,s6m,s9m,s1y,s18m,s2y,e3m,e6m,e9m,e1y,e18m,e2y,e_all= Fit_BF_sigmalow(OBJ, INPUT, GUESS, S, T_ls, Rd_ls, Rf_ls)
    
    show_x=[np.round(x,4) for x in res_lsq_x]
    print('\nShow Errors...')
    print(f'PARAM={show_x}\nerr3m=\t{e3m:.3f} \terr6m=\t{e6m:.3f} \terr9m=\t{e9m:.3f} \nerr1y=\t{e1y:.3f} \terr1f=\t{e18m:.3f} \terr2y=\t{e2y:.3f} \terrall=\t{e_all:.3f}')

    GUESS = res_lsq_x
    
    return pd.Series({
        # sig0
        'sig0_3m':res_lsq_x[0+3*0],
        'sig0_6m':res_lsq_x[0+3*1],
        'sig0_9m':res_lsq_x[0+3*2],
        'sig0_1y':res_lsq_x[0+3*3],
        'sig0_18m':res_lsq_x[0+3*4],
        'sig0_2y':res_lsq_x[0+3*5],
        

        # sig1
        'sig1_3m':res_lsq_x[1+3*0],
        'sig1_6m':res_lsq_x[1+3*1],
        'sig1_9m':res_lsq_x[1+3*2],
        'sig1_1y':res_lsq_x[1+3*3],
        'sig1_18m':res_lsq_x[1+3*4],
        'sig1_2y':res_lsq_x[1+3*5],

        # kap
        'kappa_3m':res_lsq_x[2+3*0],
        'kappa_6m':res_lsq_x[2+3*1],
        'kappa_9m':res_lsq_x[2+3*2],
        'kappa_1y':res_lsq_x[2+3*3],
        'kappa_18m':res_lsq_x[2+3*4],
        'kappa_2y':res_lsq_x[2+3*5],
        
        # lamb
        'alpha':res_lsq_x[POS_ALPHA],
        'scale':res_lsq_x[POS_SCALE],
        
        # calib quotes
        'sS10P3m':s3m[0],'sS25P3m':s3m[1],'sSATM3m':s3m[2],'sS25C3m':s3m[3],'sS10C3m':s3m[4],
        'sS10P6m':s6m[0],'sS25P6m':s6m[1],'sSATM6m':s6m[2],'sS25C6m':s6m[3],'sS10C6m':s6m[4],
        'sS10P9m':s9m[0],'sS25P9m':s9m[1],'sSATM9m':s9m[2],'sS25C9m':s9m[3],'sS10C9m':s9m[4],
        'sS10P1y':s1y[0],'sS25P1y':s1y[1],'sSATM1y':s1y[2],'sS25C1y':s1y[3],'sS10C1y':s1y[4],
        'sS10P18m':s18m[0],'sS25P18m':s18m[1],'sSATM18m':s18m[2],'sS25C18m':s18m[3],'sS10C18m':s18m[4],
        'sS10P2y':s2y[0],'sS25P2y':s2y[1],'sSATM2y':s2y[2],'sS25C2y':s2y[3],'sS10C2y':s2y[4],
        
        # err
        'me3m':e3m,
        'me6m':e6m,
        'me9m':e9m,
        'me1y':e1y,
        'me18m':e18m,
        'me2y':e2y,

        # error total
        'me':e_all,       

    })
```
